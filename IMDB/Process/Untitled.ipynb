```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
result = requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=user_rating')
# result2 = requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt')
# result3 = requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=101&ref_=adv_nxt')
# result4 = requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=151&ref_=adv_nxt')
# result5 = requests.get('https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=201&ref_=adv_nxt')

src = result.content
soup = BeautifulSoup(src, 'lxml')
```

```python
df = pd.DataFrame()
```

```python
# TITLE
images = soup.find_all('img', class_='loadlate') #['alt'], berisi title
title = []
for image in images:
    title.append(image['alt'])
len(title)
df['Title'] = title
```

```python
# DESCRIPTION
description = []
descs = soup.find_all('p', class_='text-muted')
for i in range(len(descs)):
    if (i % 2) != 0:
        description.append(descs[i].text)
df['Description'] = description
df['Description'] = df['Description'].apply(lambda ser: ser[5:])
```

```python
# RUNTIME, GENRE
runtimes = []
genres = []
for i in range(len(descs)):
    if (i % 2) == 0:
        for runtime in descs[i].find_all('span', class_='runtime'):
            runtimes.append(runtime.text)
        for genre in descs[i].find_all('span', class_='genre'):
            genres.append(genre.text)
df['Runtime'] = runtimes
df['Genre'] = genres
```

```python
df['Runtime'] = df['Runtime'].apply(lambda ser: int(ser[:3]))
df['Genre'] = df['Genre'].apply(lambda ser: ser[1:])
```

```python
# YEAR
years = []
for year in soup.find_all('span', class_='lister-item-year'):
    years.append(year.text)
df['Year'] = years
```

```python
df['Year'] = df['Year'].apply(lambda ser: ser.replace('(', '').replace(')', ''))
# df['Year'].loc[2] = '2019'
# df['Year'] = df['Year'].astype('int64')
```
